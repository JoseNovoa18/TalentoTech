```python
# Ejercicio 2: Calcular el total de ventas por región

ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
```

```python
# Ejercicio 3: Filtrar el DataFrame para mostrar solo las ventas realizadas por David

ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
```

```python
# Ejercicio 4: Calcular el promedio por región

promedio_por_region = df.groupby('Region')['Ventas'].mean()
print(promedio_por_region)
```

```python
# Ejercicio 5: Calcular el promedio de ventas por región con dos decimales

ventas_por_producto = df.groupby('Producto')['Ventas'].mean().round(2)
print(ventas_por_producto)
```

```python
# Ejercicio 6: Regiones con mayores ventas de mayor a menor

top_3_regiones = df.groupby('Region')['Ventas'].sum().nlargest(3)

print(top_3_regiones)
```

```python
# Ejercicio 7: Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor

ventas_mayores_10000 = df[df['Ventas'] > 10000].sort_values(by='Ventas', ascending=False)
print(ventas_mayores_10000)
```

```python
# Ejercicio 8: Ordenar el DataFrame por ventas en orden descendente
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

```python
# Ejericicio 9: Eliminar la columna "Vendedor"
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
```
